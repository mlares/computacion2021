```python
import math as m
import matplotlib.pyplot as plt
```

```python
def sFibo(N):
  F=[1,1]
  for n in range(N-1):
    F.append(F[n]+F[n+1])
  return F
```

```python
F=sFibo(20)
n=len(F)-1; x=n*[0]
for i in range(n):
  x[i]=F[i+1]/F[i]
print(x)
print(x[-1])
print((1+m.sqrt(5))/2)
```

```python
plt.figure(figsize=(18, 10))
plt.plot(x, marker='o', color='teal')
plt.title("x")
plt.grid()
plt.show()
```
