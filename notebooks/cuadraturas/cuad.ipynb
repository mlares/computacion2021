```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f(x):
    return np.sin(3*x) + x**2
```

```python
xd = np.array([1, 2, 3])
yd = f(xd)
```

```python
N = 100
t = np.linspace(xd.min(), xd.max(), N)
y0 = np.repeat(0, N)
```

```python
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot()

ax.plot(t, f(t), color='navy', linestyle='-')
ax.plot(xd, yd, 'o', mfc='white', mec='navy', markersize=10, mew=2)
ax.set_ylim(bottom=0)
ax.grid(alpha=0.3)
ax.fill_between(t, y0, f(t), color='cornflowerblue', alpha=0.2)
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('f(x)', fontsize=20)
```

```python
from interp_tools import interp_lagrange
from scipy.interpolate import interp1d
```

```python
f1 = interp1d(xd, yd, kind='nearest')
f2 = interp1d(xd, yd, kind='previous')
f3 = interp1d(xd, yd, kind='next')
f4 = interp1d(xd, yd)
#f5 = interp1d(xd, yd, kind='cubic')
```

```python
# Color schemes

c1 = '#00203FFF'
c2 = '#ADEFD1FF'

c1 = '#A07855FF'
c2 = '#D4B996FF'
ce1 = 'k'
ce2 = 'peru'
```

```python
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot()

ax.plot(t, f(t), color=ce1, linestyle='-')
ax.plot(xd, yd, 'o', mfc='white', mec=ce1, markersize=10, mew=2)
ax.set_ylim(bottom=0)
ax.grid(alpha=0.3)
ax.fill_between(t, y0, f(t), color=c2, alpha=0.2)
ax.plot(t, f1(t), color=ce2)

ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('f(x)', fontsize=20)
```

```python
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot()

ax.plot(t, f(t), color='k', linestyle='-')
ax.plot(xd, yd, 'o', mfc='white', mec='k', markersize=10, mew=2)
ax.set_ylim(bottom=0)
ax.grid(alpha=0.3)
ax.fill_between(t, y0, f(t), color=c1, alpha=0.2)

ax.plot(t, f1(t), color=ce2)
ax.fill_between(t, y0, f1(t), color=c2, alpha=0.2)

ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('f(x)', fontsize=20)
```

```python
for func in [f1, f2, f3,f4]:

    fig = plt.figure(figsize=(9, 9))
    ax = fig.add_subplot()

    ax.plot(t, f(t), color='k', linestyle='-')
    ax.plot(xd, yd, 'o', mfc='white', mec='k', markersize=10, mew=2)
    ax.set_ylim(bottom=0)
    ax.grid(alpha=0.3)
    ax.fill_between(t, y0, f(t), color=c1, alpha=0.2)

    ax.plot(t, func(t), color=ce2)
    ax.fill_between(t, y0, func(t), color=c2, alpha=0.2)

    ax.set_xlabel('x', fontsize=20)
    ax.set_ylabel('f(x)', fontsize=20)
```
