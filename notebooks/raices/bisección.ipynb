```python
def bisección(f, a, b, tolerancia_x, tolerancia_y, tolerancia_N):
    """
    Calcula una aproximación a una raíz de f por el método de bisección
    
    Parameters
    ----------
    f: function
    a: float
       limite inferior
    b: float
       limite superior
    tolerrancia_x: float
    tolerrancia_f: float
    tolerrancia_N: int
    
    returns
    -------
    x: float
       Aproximación de una raíz
    """
    condición = True
    contador = 0
    x_anterior = a
    lista_x = list()
    lista_f = list()
    lista_a = [a]
    lista_b = [b]
    while condición:
        x = (a+b)/2
        if f(x)*f(a) > 0:
            a = x
        else:
            b = x
        contador += 1
        cond1 = abs(x-x_anterior) > tolerancia_x
        cond2 = abs(f(x)-f(x_anterior)) > tolerancia_f
        cond3 = contador < tolerancia_N
        condición = (cond1 and cond2) and cond3
        x_anterior = x
        lista_x.append(x)
        lista_f.append(f(x))
        lista_a.append(a)
        lista_b.append(b)
    return lista_x, lista_f, lista_a, lista_b
```

```python
def plot_bisección(x, y, La, Lb):
    
    import numpy as np
    
    Npasos = len(x)
    x = np.array(x)
    y = np.array(y)

    fig = plt.figure(figsize=(14, 8))
    ax = fig.subplots(2, 3)

    ax[0, 0].axhline(y=x[-1], color='coral', linewidth=0.8)
    ax[0, 0].plot(x)
    ax[0, 0].set_xlabel('Número de iteraciones')
    ax[0, 0].set_ylabel('x')

    ax[1, 1].axhline(y=0, color='coral', linewidth=0.5)
    ax[1, 1].plot(range(Npasos), y)
    ax[1, 1].set_ylabel('y')
    ax[1, 1].set_xlabel('Número de iteraciones')

    ax[1, 0].axvline(x=0, color='coral', linewidth=0.5)
    ax[1, 0].axhline(y=0, color='coral', linewidth=0.5)
    ax[1, 0].plot(x-x[-1], y-y[-1], marker='o')
    ax[1, 0].set_ylabel('error en y')
    ax[1, 0].set_xlabel('error en x')

    ax[0, 1].plot(x, color='coral')
    for i, intervalo in enumerate(zip(La, Lb)):
        ax[0, 1].plot([i, i], [intervalo[0], intervalo[1]], color='teal', linewidth=3)
        ax[0, 1].plot([0, Npasos], [x[-1], x[-1]], '--', color='grey')
    ax[0, 1].set_ylabel('Intervalo [a, b]')
    ax[0, 1].set_xlabel('Número de iteraciones')
    
    ax[0, 2].axhline(y=0, color='coral', linewidth=0.5)
    ax[0, 2].plot(abs(x - x[-1]))
    ax[0, 2].set_ylabel('módulo error en x')
    ax[0, 2].set_xlabel('Número de iteraciones')
    ax[0, 2].set_yscale('log')
                  
    ax[1, 2].axhline(y=0, color='coral', linewidth=0.5)
    ax[1, 2].plot(abs(y- y[-1]))
    ax[1, 2].set_ylabel('módulo error en y')
    ax[1, 2].set_xlabel('Número de iteraciones')
    ax[1, 2].set_yscale('log')    
```
