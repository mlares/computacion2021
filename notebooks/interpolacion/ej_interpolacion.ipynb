```python
def fa(x, n):
    n = 6
    aprox = 0
    for k in range(n):
        a_k = (-1)**k * (x-1)**k
        aprox = aprox + a_k  
    return aprox
```

```python
x = np.linspace(0.2, 3, 200)

fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot()

plt.plot(x, f(x), linewidth=5, color='silver')
plt.plot(x, fa(x, 20), '--')
plt.axvline(1, color='burlywood', lw=0.4)
plt.plot([1], [f(1)], 'o', markersize=7, linestyle='None', mfc='white', mec='black')
plt.xlabel('x')
plt.ylabel('y')
fig.savefig('taylor.pdf')
```
