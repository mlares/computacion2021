```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import interpolate
```

```python
x1 = [-5, -4.6, -4, -1, 0, 1, 4, 4.6, 5]
y1 = [0,     5,  7,  4, 3, 4, 7,   5, 0]
x2 = [-5,-4.5,-3,  0,  3,4.5, 5]
y2 = [ 0,-4,-10,-15, -10,-4, 0]
x = x1+x2
y = y1+y2
```

```python
plt.plot(x1, y1, 'o', x2, y2, 'o')
```

```python
sup = interpolate.splrep(x1, y1, s=0)
sdn = interpolate.splrep(x2, y2, s=0)
```

```python
xs = np.linspace(-5, 5, 100)
ysup = interpolate.splev(xs, sup)
ysdn = interpolate.splev(xs, sdn)
```

```python
f = plt.figure(figsize=(5, 5))
ax = f.add_subplot()

ax.plot(x, y, 'o')
ax.plot(xs, ysup, xs, ysdn)
ax.set_xlabel('x')
ax.set_ylabel('y')
f.savefig('heart_dots.pdf')
```

```python
f = plt.figure(figsize=(5, 5))
ax = f.add_subplot()

ax.plot(x, y, 'o')
ax.plot(xs, ysup, xs, ysdn)
ax.fill_between(xs, ysdn, ysup, hatch='///', color='pink', alpha=0.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
f.savefig('heart_dots.pdf')
```
