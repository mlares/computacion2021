```python
import numpy as np
import math
from matplotlib import pyplot as plt
```

Serie de f(x)=1/x

```python
def f(x):
    return 1/x
```

```python
x = np.linspace(1, 3, 500)
```

```python
plt.plot(x, f(x))
```

```python
x = 1.4
```

```python
def fa(x, n):
    aprox = 0
    for k in range(n):
        a_k = (-1)**k * (x-1)**k
        aprox = aprox + a_k  
    return aprox
```

```python
print(x, f(x), fa(x, 6))
```

```python
x = np.linspace(0.2, 3, 200)

fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot()

plt.plot(x, f(x), linewidth=5, color='silver', label='Funcion')
plt.plot(x, fa(x, 4), '--', label='Aprox. de Taylor')
plt.axvline(1, color='burlywood', lw=0.4)
plt.plot([1], [f(1)], 'o', markersize=7, linestyle='None', mfc='white', mec='black')
plt.xlabel('x')
plt.ylabel('y')
ax.legend()
fig.savefig('taylor.pdf')
```
